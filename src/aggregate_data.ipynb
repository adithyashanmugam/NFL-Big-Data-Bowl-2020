```python
import pandas as pd
import numpy as np
import itertools as it
import matplotlib.pyplot as plt
from scipy import stats
from collections import defaultdict
import csv
```

```python
receiver_ability = pd.read_csv('../intermediate_data/receiver_ability_scaled.csv')
yac_labels = pd.read_csv('../intermediate_data/yac_labels.csv')
play_scores = pd.read_csv('../intermediate_data/play_scores_without_def_scores.csv')
tackle_scores = pd.read_csv('../intermediate_data/tacklers_per_play.csv')
closest_defender = pd.read_csv('../intermediate_data/closest_defender.csv')
```

```python
merge1 = play_scores.merge(receiver_ability, how='inner', on='nflId')
merge2 = merge1.merge(tackle_scores, how='inner', on=['gameId', 'playId'])
merge25 = merge2.merge(closest_defender, how='inner', on=['gameId', 'playId'])
merge3 = merge25.merge(yac_labels, how='inner', on=['gameId', 'playId'])
df4 = merge3.drop(columns=['nflId'])
df5 = df4.loc[(df4['play_score'] != 20)]
df4.to_csv('../intermediate_data/full_data.csv', index=False)
df4
```
