```python
week1 = pd.read_csv('./data/week1.csv')
game1 = week1[(week1['gameId'] == GAME1)]
PLAY1 = pd.unique(game1.playId)[0]
g1p1 = game1[game1['playId']==PLAY1]
pd.unique(g1p1[g1p1['event']=='pass_forward'].time)
pd.unique(g1p1[g1p1['event']=='pass_arrived'].time)
```

```python
plays = pd.read_csv('./data/plays.csv')
plays = plays[(plays['gameId'] == GAME1) & plays.passResult.isin(['C', 'I'])]
for p in pd.unique(plays.playId):
    temp_pd = plays[plays['playId'] == p].playDescription.values[0]
    split_to = temp_pd.split('to ')
    receiver = split_to[1].split()[0] if len(split_to) > 1 else ''
#     if(receiver == 'PHI'): print(temp_pd)
    print(receiver)
```
